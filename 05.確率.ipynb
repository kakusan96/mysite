```python
#stats.<確率分布>.<メソッド>

#<確率分布>
#norm : 正規分布 (Normal/Gaussian)ベル型の曲線を描く連続分布。多くの自然現象や測定誤差などを表現する。平均と標準偏差で特徴付けられる。
#uniform : 一様分布 (Uniform)指定された区間内のすべての値が等しい確率で発生する連続分布。最小値と最大値で特徴付けられる。
#randint : 離散一様分布 (Discrete Uniform)指定された整数の範囲内で、すべての値が等確率で出現する離散分布。最小値と最大値で特徴付けられる。
#binorm : 二項分布 (Binomial)n回の独立した試行において、各試行で成功する確率がpである場合の成功回数の分布。試行回数nと成功確率pで特徴付けられる。
#poisson : ポアソン分布 (Poisson)単位時間あたりに発生する事象の回数を表す離散分布。平均発生率λで特徴付けられる。希少事象のカウントに適している。
#expon : 指数分布 (Exponential)イベント間の待ち時間を表す連続分布。一定の割合でランダムに発生する事象の間隔を表現する。率パラメータλで特徴付けられる。
#beta : ベータ分布0から1の範囲の値を取る連続分布。2つの形状パラメータα、βによって特徴付けられる。割合やベイズ統計でよく使用される。
#gamma : ガンマ分布正の実数を取る連続分布。形状パラメータkとスケールパラメータθで特徴付けられる。待ち時間や正の値を持つ物理量のモデル化に使用。
#lognorm : 対数正規分布変数の対数が正規分布に従う連続分布。所得分布や株価など、正の値で右に歪んだデータによく適合する。
#t : スチューデントのt分布正規分布に似ているが、より裾の重い連続分布。自由度によって特徴付けられる。小標本統計で重要。
#chi2 : カイ二乗分布正規分布に従う確率変数の二乗和の分布。自由度によって特徴付けられる。適合度検定などに使用される。
#f : F分布2つのカイ二乗分布の比から導かれる連続分布。分散分析(ANOVA)などで使用される。
#multinomial : 多項分布二項分布の拡張で、複数のカテゴリがある場合の出現回数の同時分布。
#weibull : ワイブル分布故障率モデリングに使われる連続分布。形状パラメータとスケールパラメータで特徴付けられる。
#pareto : パレート分布べき法則に従う連続分布。所得分布、都市の人口分布などの社会現象のモデル化に使用される。
#laplace : ラプラス分布二重指数分布とも呼ばれる対称な連続分布。中心からの距離が指数関数的に減少する。
#cauchy : コーシー分布非常に裾の重い連続分布。平均や分散が存在しない特徴を持つ。
#hypergeom : 超幾何分布有限母集団からの非復元抽出における成功回数の分布。
#nbinom : 負の二項分布r回の成功が観測されるまでの試行回数の分布。
#geom : 幾何分布1回の成功が観測されるまでの試行回数の分布。負の二項分布の特殊ケース。

#<メソッド>
#rvs() : 乱数サンプルを生成 (random variates)stats.norm.rvs(loc=0, scale=1, size=10)
#pdf() : 確率密度関数 (probability density function)stats.norm.pdf(x, loc=0, scale=1)
#pmf() : 確率質量関数 (probability density function)stats.norm.pmf(low, high)
#cdf() : 累積分布関数 (cumulative distribution function)stats.norm.cdf(x, loc=0, scale=1)
#ppf() : パーセント点関数/逆CDF (percent point function)stats.norm.ppf(0.95, loc=0, scale=1)
#sf() : 生存関数/1-CDF (survival function)stats.norm.sf(x, loc=0, scale=1)
#isf() : 逆生存関数 (inverse survival function)stats.norm.isf(0.05, loc=0, scale=1)
#stats() : 分布の統計量（平均・分散・歪度・尖度）を計算stats.norm.stats(loc=0, scale=1, moments='mvsk')
#fit() : データに分布をフィットしてパラメータを推定stats.norm.fit(data)
#interval() : 信頼区間を計算stats.norm.interval(0.95, loc=0, scale=1)
#moment() : 任意の次数のモーメント（積率）を計算stats.norm.moment(n, loc=0, scale=1)
#entropy() : 分布のエントロピーを計算stats.norm.entropy(loc=0, scale=1)
#expect() : 確率変数の期待値を計算stats.norm.expect(func, loc=0, scale=1)
#median() : 分布の中央値を計算stats.norm.median(loc=0, scale=1)
#mean() : 分布の平均値を計算stats.norm.mean(loc=0, scale=1)
#var() : 分布の分散を計算stats.norm.var(loc=0, scale=1)
#std() : 分布の標準偏差を計算stats.norm.std(loc=0, scale=1)
#logpdf() : 確率密度関数の自然対数stats.norm.logpdf(x, loc=0, scale=1)
#logcdf() : 累積分布関数の自然対数stats.norm.logcdf(x, loc=0, scale=1)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import stats
```

```python
#離散一様分布
#1〜6の一様分布をランダムにデータを生成
stats.randint.rvs(1, 7)
```

```python
#連続型一様分布
#0〜1の間でランダムな数値を返してくれる
stats.uniform.rvs(0, 1)
```

```python
#離散型確率分布→確率質量関数
#連続型確率分布→確率密度関数
#確率分布の描画

#連続値のX軸の作成
#-3〜3を100個に区切る
#x = np.linspace(-3, 3, 100)
#離散型のx軸
#-3〜3までの離散値
#x = np.arange(-3, 4)
x
```

```python
#確率変数が1〜6をとる離散型一様分布の確率分布
x = np.arange(1, 7)
y = stats.randint(1, 7).pmf(x)
plt.plot(x, y, "o")
plt.vlines(x, 0, y)
```

```python
#確率変数が0〜1をとる連続型一様分布の確率分布
x = np.linspace(-2, 2, 100)
y = stats.uniform(loc=0, scale=1).pdf(x)
plt.plot(x, y)
#確率密度関数が描く面積が確率になる。なのである値からある値までの確率変数をとってその面積を求めると確率になる。
#確率密度関数が描く全面積は1になる。

#累積分布関数
#確率密度関数を負の無限大からある値まで確率を積み上げた値を返す関数。ある値以下の確率変数を取る確率を返す。
y2 = stats.uniform(loc=0, scale=1).cdf(x)
plt.plot(x, y2)
```

```python
#正規分布の確率密度関数と累積分布関数を描画する。
#loc→平均 scale=分散
x = np.linspace(-10, 10, 100)
y = stats.norm(loc=0, scale=2).pdf(x)
y2 = stats.norm(loc=0, scale=2).cdf(x)
y3 = stats.norm(loc=0, scale=2).sf(x)
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
#カーネル密度推定
#観測した分布から確率密度関数を推定する
df = sns.load_dataset("tips")
x = np.linspace(np.min(df["total_bill"]), np.max(df["total_bill"]), 100)
y = stats.gaussian_kde(df["total_bill"]).pdf(x)
#plt.plot(x, y)
#ヒストグラムと推定した確率密度関数を描画
sns.displot(df["total_bill"], kde = True)
sns.displot(df["total_bill"], kind="ecdf")
```
